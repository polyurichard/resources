```python
from img2table.ocr import TesseractOCR, PaddleOCR
from img2table.document import PDF, Image
import pandas as pd


# Instantiation of OCR
tesseractocr = TesseractOCR(n_threads=1, lang="eng")
paddle_ocr = PaddleOCR(lang="en", kw={"use_dilation": True})

def extract_doc_demo(doc, ocr = tesseractocr):
    # Table extraction
    result = doc.extract_tables(ocr=ocr,
                                        implicit_rows=False,
                                        borderless_tables=True,
                                        min_confidence=50)



    print(result)

    for page, tables in result.items():
        print("This is Page", page)

        # iterate over the tables in one page
        for table in tables:
        # access the table at the cell-level
            display(table.df)
            # for id_row, row in enumerate(table.content.values()):
            #     for id_col, cell in enumerate(row):

            #         x1 = cell.bbox.x1
            #         y1 = cell.bbox.y1
            #         x2 = cell.bbox.x2
            #         y2 = cell.bbox.y2
            #         value = cell.value
            #         print(value)

def extract_image_demo(image, ocr = tesseractocr):
    # Table extraction
    result = image.extract_tables(ocr=ocr,
                                        implicit_rows=False,
                                        borderless_tables=True,
                                        min_confidence=50)



    print(result)

    for page, table in enumerate(result):
        print("This is Page", page)

        display(table.df)
```
